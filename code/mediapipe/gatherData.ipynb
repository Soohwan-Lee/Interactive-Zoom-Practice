# 3. Train Custom Model Using Scikit Learn

## 3.1 Read in Collected Data and Process

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('coords.csv')
```

```python
df = df.dropna(axis=0)
```

```python
df.head()
```

```python
df.tail()
```

```python
df[df['class']=='Background']
```

```python
X = df.drop('class', axis=1) # features
y = df['class'] # target value
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25, random_state=1234)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

## 3.2 Train Machine Learning Classification Model

```python
from sklearn.pipeline import make_pipeline 
from sklearn.preprocessing import StandardScaler 

from sklearn.linear_model import LogisticRegression, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
pipelines = {
    'lr':make_pipeline(StandardScaler(), LogisticRegression()),
    'rc':make_pipeline(StandardScaler(), RidgeClassifier()),
    'rf':make_pipeline(StandardScaler(), RandomForestClassifier()),
    'gb':make_pipeline(StandardScaler(), GradientBoostingClassifier()),
    'sv':make_pipeline(StandardScaler(), SVC()),
}
```

```python
fit_models = {}
for algo, pipeline in pipelines.items():
    model = pipeline.fit(X_train, y_train)
    fit_models[algo] = model
```

```python
fit_models
```

```python
fit_models['rf'].predict(X_test)
```

## 3.3 Evaluate and Serialize Model 

```python
from sklearn.metrics import accuracy_score # Accuracy metrics 
import pickle 
```

```python
for algo, model in fit_models.items():
    pred = model.predict(X_test)
    print(algo, accuracy_score(y_test, pred))
```

```python
fit_models['sv'].predict(X_test)
```

```python
with open('body_language.pkl', 'wb') as f:
    pickle.dump(fit_models['rf'], f)
```
